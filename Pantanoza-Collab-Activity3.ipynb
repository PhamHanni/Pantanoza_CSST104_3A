**Exercise 1**

1. Importing Libraries

```python
import pandas as pd #data manipulation
import matplotlib.pyplot as plt #data visualization
```

2. Loading Datasets

```python
data = pd.read_csv('sales_data_2.csv', encoding = 'latin-1')
```

3. Data Loading and Initial Analysis

```python
statistics = data.describe()
```

4. Data Manipulation

```python
prod_rev = data.groupby('Product')['Revenue'].sum()
high_rev_prod = prod_rev.idxmax()
avg_quan_pd = data['Quantity Sold'].mean()

print(prod_rev)
print(high_rev_prod)
print(avg_quan_pd)
```

5. Data Visualization

```python
prod_rev.plot(kind = 'bar', color = 'pink')
plt.xlabel('Product')
plt.ylabel('Revenue')
plt.title('Revenue by Product')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

6. Analysis Report

```python
report = """
Analysis Report

1. Data Loading and Initial Analysis :
   Basic statistics of the  dataset :
   {}
2. Data Manipulation :
   Total revenue for each product :
   {}
   Product with the highest revenue : {}
   Average quantity sold per day : {:.2f}
3. Data Visualization :
   Bar Chart displaying revenue for each product is attached.
4. Insights :
   -The product {} generates the highest revenue.
   - On average, {:.2f} units sold per day.

""".format(statistics, prod_rev, high_rev_prod, avg_quan_pd, high_rev_prod ,avg_quan_pd)

print(report)
```

7. Saving the report to a file

```python
with open('report.txt', 'w') as f:
  f.write(report)
```
