# **Creating and Testing a Non-Deterministic Finite Automata (NFA) Convert to Deterministic Finite Automata (DFA) in Python using automata-lib**

**1. Install the latest version of Automata Library**

```python
pip install automata-lib
```

**2. Import the NFA class from the automata.fa.nfa module.**

```python
from automata.fa.nfa import NFA
```

**3. Define the NFA with it's states.*** **states**
* **input symbols**
* **transitions**
* **initial states**
* ***final states***

COLLAB ACT1.png

```python
nfa_example = NFA(
    states={'q0', 'q1'},
    input_symbols={'0', '1'},
    transitions={
        'q0': {'0': {'q0', 'q1'}, '1': {'q0', 'q1'}},
        'q1': {'1': {'q0', 'q1'}},
    },
    initial_state='q0',
    final_states={'q1'}
)
```

```python
def is_accepted_example(input_str):
    return nfa_example.accepts_input(input_str)
```

```python
test_strings = ['0011', '00011', '001110011', '00111', '0001110011']

for test_str in test_strings:
    if is_accepted_example(test_str):
        print(f'String "{test_str}" is accepted.')
    else:
        print(f'String "{test_str}" is not accepted.')
```

**4. Import the DFA class from the automata.fa.dfa module.**

```python
from automata.fa.dfa import DFA
```

**5. Convert NFA to DFA**

```python
dfa_example = DFA.from_nfa(nfa_example)
```

**3. Define the DFA with it's states.*** **states**
* **input symbols**
* **transitions**
* **initial states**
* ***final states***

```python
print("DFA States:", dfa_example.states)
print("DFA Input Symbols:", dfa_example.input_symbols)
print("DFA Transitions:", dfa_example.transitions)
print("DFA Initial State:", dfa_example.initial_state)
print("DFA Final States:", dfa_example.final_states)
```

**4. Test the DFA with a list of strings**

```python
test_strings = ['0011', '00011', '001110011', '00111', '0001110011']

for test_str in test_strings:
    if is_accepted_example(test_str):
        print(f'String "{test_str}" is accepted.')
    else:
        print(f'String "{test_str}" is not accepted.')
```
